# Exploring Convolutional Layers Through Data and Experiments

The objective of this notebook is to understand and apply convolutional neural network architecture, analyzing the impact of its hyperparameters and its advantage over dense layers through data analysis and technical communication of results by experimenting with public data.

## Dataset Selection



The OCTMNIST dataset is appropriate for convolutional layers because it is composed of 2D image data with strong spatial structure. Each sample is a retinal OCT image in which neighboring pixels are highly correlated and form meaningful local patterns such as retinal layers, edges, and pathological textures.

Convolutional layers are well suited for this type of data, as they exploit local receptive fields, weight sharing, and spatial hierarchy preservation, allowing the model to efficiently learn visual features that are relevant for medical image classification.

In addition, OCTMNIST is already integrated and readily accessible in Python, and it comes pre-split into training, validation, and test sets, which simplifies the experimental pipeline and reduces the risk of data leakage.

The dataset is also computationally lightweight, since all images are normalized and resized to 2828 pixels. This enables fast training and efficient memory usage on standard laptops or cloud notebooks, while still preserving enough structural information for convolutional neural networks to learn discriminative features.
